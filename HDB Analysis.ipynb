```python
#residual analysis

# adjust the figure size
plt.figure(figsize=(10,8))

# plotting residual errors in training data
plt.scatter(reg.predict(X_train), reg.predict(X_train) - y_train, c = 'b', s = 40, label = 'Train data', alpha = 0.5)

# plotting residual errors in test data
plt.scatter(reg.predict(X_test), reg.predict(X_test) - y_test, c = 'r', s = 40, label = 'Test data')

# plotting line for zero residual error
plt.hlines(y = 0, xmin = -100, xmax = 1000, linewidth = 3)

# plotting legend
plt.legend(loc = 'upper right')

# plot title
plt.title("Residual errors plot")

# function to show plot
plt.show()
```

```python
#QQ Plot
import pylab
import scipy.stats as stats

for var in X_train.columns[1:]:
    plt.figure(figsize=(15,6))

    plt.subplot(1, 2, 1)
    X_train[var].hist()
    plt.title('Distribution of '+ var)

    plt.subplot(1, 2, 2)
    stats.probplot(X_train[var], dist="norm", plot=pylab)
    plt.title('Q-Q plot of '+ var)

    plt.show() 
```
